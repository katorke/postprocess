```python
# Import libraries
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# function to read your COLVAR files
def read_colvar(file_name):
    with open(file_name) as f:
        first_line = f.readline()
    fl = first_line[10:-1]
    names = fl.split()
    df = pd.read_csv(file_name,delim_whitespace=True, names=names, comment='#')
    return df
```

```python
# load your colvar files
rf = read_colvar('R5tide_forward/COLVAR')
rr = read_colvar('R5tide_reverse/COLVAR')
cf = read_colvar('chen_forward/COLVAR')
cr = read_colvar('chen_reverse/COLVAR')
```

```python
# just checking which columns I have
rf.columns
```

```python
# Plots of CVs
# I use these to look at which structures I should pick out for setting up multiwalker simulations
# but could also be useful for loading multiple different COLVAR files/comparing/etc. 

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=[15,10])
rf.plot(ax=axes[0,0],x='time',y='IONCL1', title='RF')
rr.plot(ax=axes[0,1],x='time',y='IONCL1',title='RR')
cf.plot(ax=axes[1,0],x='time',y='IONCL1',title='CF')
cr.plot(ax=axes[1,1],x='time',y='IONCL1',title='CR')
```
