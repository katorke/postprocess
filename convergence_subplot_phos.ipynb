```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from each walker
cvvals0 = np.loadtxt('phos/COLVAR.0',comments='#')
cvvals1 = np.loadtxt('phos/COLVAR.1',comments='#')
cvvals2 = np.loadtxt('phos/COLVAR.2',comments='#')
cvvals3 = np.loadtxt('phos/COLVAR.3',comments='#')
cvvals4 = np.loadtxt('phos/COLVAR.4',comments='#')
```

```python
# This is just cleaning up the number of data points a bit
new0=cvvals0[0::1000]
new1=cvvals1[0::1000]
new2=cvvals2[0::1000]
new3=cvvals3[0::1000]
new4=cvvals4[0::1000]
```

```python
# Plot
fig, axs = plt.subplots(5,figsize=(15,12))
axs[0].set_title('Mutant 1: Phosphate',fontsize=18)
axs[0].plot(new0[:,0]/1000,new0[:,3],color='k')
axs[0].plot(new1[:,0]/1000,new1[:,3],color='k',alpha=0.2)
axs[0].plot(new2[:,0]/1000,new2[:,3],color='k',alpha=0.2)
axs[0].plot(new3[:,0]/1000,new3[:,3],color='k',alpha=0.2)
axs[0].plot(new4[:,0]/1000,new4[:,3],color='k',alpha=0.2)
axs[0].axhline(y=1.75,color='r')
axs[0].set_ylabel('Distance (nm)',fontweight='bold')

axs[1].plot(new0[:,0]/1000,new0[:,3],color='k',alpha=0.2)
axs[1].plot(new1[:,0]/1000,new1[:,3],color='k')
axs[1].plot(new2[:,0]/1000,new2[:,3],color='k',alpha=0.2)
axs[1].plot(new3[:,0]/1000,new3[:,3],color='k',alpha=0.2)
axs[1].plot(new4[:,0]/1000,new4[:,3],color='k',alpha=0.2)
axs[1].axhline(y=1.75,color='r')
axs[1].set_ylabel('Distance (nm)',fontweight='bold')

axs[2].plot(new0[:,0]/1000,new0[:,3],color='k',alpha=0.2)
axs[2].plot(new1[:,0]/1000,new1[:,3],color='k',alpha=0.2)
axs[2].plot(new2[:,0]/1000,new2[:,3],color='k')
axs[2].plot(new3[:,0]/1000,new3[:,3],color='k',alpha=0.2)
axs[2].plot(new4[:,0]/1000,new4[:,3],color='k',alpha=0.2)
axs[2].axhline(y=1.75,color='r')
axs[2].set_ylabel('Distance (nm)',fontweight='bold')

axs[3].plot(new0[:,0]/1000,new0[:,3],color='k',alpha=0.2)
axs[3].plot(new1[:,0]/1000,new1[:,3],color='k',alpha=0.2)
axs[3].plot(new2[:,0]/1000,new2[:,3],color='k',alpha=0.2)
axs[3].plot(new3[:,0]/1000,new3[:,3],color='k')
axs[3].plot(new4[:,0]/1000,new4[:,3],color='k',alpha=0.2)
axs[3].axhline(y=1.75,color='r')
axs[3].set_ylabel('Distance (nm)',fontweight='bold')

axs[4].plot(new0[:,0]/1000,new0[:,3],color='k',alpha=0.2)
axs[4].plot(new1[:,0]/1000,new1[:,3],color='k',alpha=0.2)
axs[4].plot(new2[:,0]/1000,new2[:,3],color='k',alpha=0.2)
axs[4].plot(new3[:,0]/1000,new3[:,3],color='k',alpha=0.2)
axs[4].plot(new4[:,0]/1000,new4[:,3],color='k')
axs[4].axhline(y=1.75,color='r')
axs[4].set_xlabel('Time (ns)',fontsize=16,fontweight='bold')
axs[4].set_ylabel('Distance (nm)',fontweight='bold')
```
