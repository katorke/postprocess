```python
# FOR METAD 2D FES
%matplotlib inline

fesdata = np.genfromtxt('fes_10.dat',comments='#');

dim1=107 # might need to change these
dim2=222 # might need to change these
fesdata = fesdata[:,0:3]
enertokcal=4.184

X=np.reshape(fesdata[:,0],[dim1,dim2],order="F")
Y=np.reshape(fesdata[:,1],[dim1,dim2],order="F") 
Z=np.reshape((fesdata[:,2]-np.min(fesdata[:,2]))/enertokcal,[dim1,dim2],order="F")  #convert to kcal/mol

spacer=1.0
lines=15
levels=np.linspace(0,lines*spacer,num=(lines+1),endpoint=True)

fig=plt.figure(figsize=(12,10)) 
axes = fig.add_subplot(111)

plt.contourf(X, Y, Z, levels, cmap="RdYlBu")
cbar = plt.colorbar(format='%.01f')
plt.xlabel('Radius of Gyration (nm)', fontsize=20)
plt.ylabel('Alphabeta', fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
axes.set_ylim([13,33.5]) # these will probs change
axes.set_xlim([0.4,1.1]) # these will probs change
plt.show()
```
