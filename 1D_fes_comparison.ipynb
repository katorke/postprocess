```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the two final fes files from plumed
# plot
fesdatno = np.loadtxt('nophos/fes_10.dat',skiprows=5)
fesdatphos = np.loadtxt('phos/fes_10.dat',skiprows=5)
plt.figure(figsize=(8,5))
plt.plot(fesdatno[:,0],fesdatno[:,1],label='Chlorine', c='green', linestyle='-')
plt.plot(fesdatphos[:,0],fesdatphos[:,1],label='Phosphate', c='blue', linestyle='-')
plt.xlabel('Distance (nm)',fontsize=18)
plt.ylabel('Free Energy (kJ/mol)', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim([-1, 30])
plt.xlim([0,4.5])
plt.legend(fontsize=14)
plt.title('Forward Analog', fontsize=18)
```
