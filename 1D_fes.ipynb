```python
# import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load all the data, then plot to see difference between 100%/90%/etc.
f2, ax2 = plt.subplots(1,1, figsize=(10,7))
kbT=2.49
last_num = 10
fes = pd.read_csv('fes_10.dat', comment='#', names=['d','projection', 'ignore'], delim_whitespace=True)
fes_9 = pd.read_csv('fes_'+str(int(last_num*0.9))+'.dat', comment='#', names=['d','projection','ignore'], delim_whitespace=True)
fes_8 = pd.read_csv('fes_'+str(int(last_num*0.8))+'.dat', comment='#', names=['d','projection','ignore'], delim_whitespace=True)
fes_7 = pd.read_csv('fes_'+str(int(last_num*0.7))+'.dat', comment='#', names=['d','projection','ignore'], delim_whitespace=True)
fes_6 = pd.read_csv('fes_'+str(int(last_num*0.6))+'.dat', comment='#', names=['d','projection','ignore'], delim_whitespace=True)
ax2.plot(fes.d,fes.projection,label='100%', c='green', linestyle='-')
ax2.plot(fes_9.d,fes_9.projection,label='90%', c='green', linestyle='-', alpha=0.8)
ax2.plot(fes_8.d,fes_8.projection,label='80%', c='green', linestyle='-', alpha=0.6)
ax2.plot(fes_7.d,fes_7.projection,label='70%', c='green', linestyle='-', alpha=0.4)
ax2.plot(fes_6.d,fes_6.projection,label='60%', c='green', linestyle='-', alpha=0.2)
ax2.set_xlabel('Distance')
ax2.set_ylabel('Free Energy')
ax2.set_xlim(0,4.5)
ax2.set_ylim(-1,50)
ax2.legend()
plt.show()
```
